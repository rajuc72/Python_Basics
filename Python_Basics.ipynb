```python
from re import A
#1.Write a Python program to print "Hello, World!
print("\"Hello, World!")

#2. Write a Python program that displays your name and age
name=input("Enter your name")
age=int(input("Enter your age"))
print("Hello my name is", name,",I am ",age,"years old")

#3.Write code to print all the pre-defined keywords in Python using the keyword library
help('keywords') # direct way
list1 = keyword.kwlist # ssigning into a variable
print(list1)

#4. Write a program that checks if a given word is a Python keyword
import keyword
list1 = keyword.kwlist
test_word = input("Enter a word to check if it's a Python keyword: ")
if test_word in list1:
    print(test_word, "is a Python keyword.")
else:
    print(test_word, "is not a Python keyword.")

#5.Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each
list1=[1,2,3,4,5]
print("original list:",list1)
list1[2]=38
print("modified list",list1)
tuple1=(1,2,3.5,4,b,c)
print("Original Tuple",tuple1)
tuple1(1)='a' #SyntaxError: cannot assign to function call here. Maybe you meant '==' instead of '='?
print("modified tuple:",tuple1) #'tuple1' is not defined

#6. Write a function to demonstrate the behavior of mutable and immutable arguments
original_list = [1, 2, 3]
original_int = 50
print("Original list:", original_list)
print("Original integer:", original_int)
def modify_list(original_list):
    original_list.append(100)
    print("Inside modify_list after modification:", original_list)

def modify_integer(original_int):
    original_int=original_int+100
    print("Inside modify_integer after modification:", original_int)

modify_list(original_list)
modify_integer(original_int)
print("After modify_list:", original_list)
print("After modify_integer:", original_int)

#7. Write a function to demonstrate the behavior of mutable and immutable arguments
original_list = [1, 2, 3]
original_int = 50
print("Original list:", original_list)
print("Original integer:", original_int)
def modify_list(original_list):
    original_list.append(100)
    print("Inside modify_list after modification:", original_list)

def modify_integer(original_int):
    original_int=original_int+100
    print("Inside modify_integer after modification:", original_int)

modify_list(original_list)
modify_integer(original_int)
print("After modify_list:", original_list)
print("After modify_integer:", original_int)

#8.Write a program to demonstrate the use of logical operators
x=int(input("Enter x value"))
y=int(input("Enter y value"))
def check_values(a, b):
    # Using logical AND
    if a > 0 and b > 0:
        print("Both a and b are positive numbers.")
    elif a > 0 and b <= 0:
        print("a is positive, but b is not.")
    elif a <= 0 and b > 0:
        print("b is positive, but a is not.")
    else:
        print("Both a and b are non-positive numbers.")

    # Using logical OR
    if a > 0 or b > 0:
        print("At least one of a or b is positive.")
    else:
        print("Neither a nor b is positive.")

    # Using logical NOT
    if not a:
        print("a is zero.")
    if not b:
        print("b is zero.")
check_values(x,y)


#9. Write a Python program to convert user input from string to integer, float, and boolean types
x=input("enter a value ")
print("Entered string is:", x)
if x.isdigit():
  a=int(x)
  print("Integer value of the given string is:", a)
else:
  print("the given string not a valid input to convert it into integer")
if x.isdigit():
  a=float(x)
  print("Float value of the given string is:", a)
else:
  print("the given string not a valid input to convert it into float")

y=bool(x)
print("Boolean value of the given string is:", y)


#10.Write code to demonstrate type casting with list elements
def type_casting_demo():
    string_list = ["1", "2", "3.5", "4.7", "5"]

    # Convert strings to integers (where applicable)
    int_list = [int(x) for x in string_list if x.isdigit()]
    print("List of integers:", int_list)

    # Convert strings to floats
    float_list = [float(x) for x in string_list if isinstance(x, str)]
    print("List of floats:", float_list)

    # Convert integers back to strings
    int_to_string_list = [str(x) for x in int_list]
    print("List of strings from integers:", int_to_string_list)

    # Type casting mixed types (int, float, string)
    mixed_list = [1, "2", 3.5, "4", 5.6]
    casted_mixed_list = [int(x) if isinstance(x, str) and x.isdigit() else float(x) if isinstance(x, str) else x for x in mixed_list]
    print("Casted combined list:", casted_mixed_list)
type_casting_demo()

#11. Write a program that checks if a number is positive, negative, or zero
a=int(input("Enter a number"))
if a>0:
  print("Number is positive")
elif a<0:
  print("Number is negative")
else:
  print("Number is zero")


#12. Write a for loop to print numbers from 1 to 10
for i in range(1,11):
  print(i,end=" ")

#13. Write a Python program to find the sum of all even numbers between 1 and 50
sum=0
total=0
print("First 50 even numbers are:")

for i in range(1,51):
  if i%2==0:
    sum=sum+i
    print(i,end=" ")
  else:
    total=total+i
print("\nSum of first 50 even numbers is:",sum)
print("Sum of first 50 odd numbers is:",total)

#14.Write a program to reverse a string using a while loop
str_name=input("Enter a string")
print("Original string is:",str_name)
print("Reversed string is:",end="")
while len(str_name)>0:
  print(str_name[-1],end="")
  str_name=str_name[:-1]

#15.Write a Python program to calculate the factorial of a number provided by the user using a while loop.
fact=1
num=int(input("Enter a number"))
while num>0:
  fact=fact*num
  num=num-1
print("Factorial of the given number is:",fact)
```
